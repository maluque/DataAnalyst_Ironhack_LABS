```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
from sklearn import datasets
# from sklearn.preprocessing import StandardScaler

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns = data["feature_names"])
# X = StandardScaler().fit_transform(X)

y = pd.Series(data["target"])

X.head()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 8)

kmeans.fit(X)

y1_pred = kmeans.predict(X)

silhouette_score(X, y1_pred)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 5)

kmeans.fit(X)

y2_pred = kmeans.predict(X)

silhouette_score(X, y2_pred)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 3)

kmeans.fit(X)

y3_pred = kmeans.predict(X)

silhouette_score(X, y3_pred)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
modelo = KMeans()

visual = KElbowVisualizer( modelo, k=(2, 15), metric='distortion')
visual.fit(X)

visual.poof();
```

```python
modelo = KMeans()

visual = KElbowVisualizer(modelo, k = (2, 15), metric = 'silhouette')

visual.fit(X)

visual.poof();
```

```python
modelo = KMeans()

visual = KElbowVisualizer(modelo, k = (2, 15), metric = 'calinski_harabasz')

visual.fit(X)

visual.poof();
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
modelo_agg = AgglomerativeClustering()

visual_agg = KElbowVisualizer(modelo_agg, k = (2, 15), metric = 'distortion')

visual_agg.fit(X)

visual_agg.poof();
```

```python
modelo_agg = AgglomerativeClustering()

visual_agg = KElbowVisualizer(modelo_agg, k = (2, 15), metric = 'silhouette')

visual_agg.fit(X)

visual_agg.poof();
```

```python
modelo_agg = AgglomerativeClustering()

visual_agg = KElbowVisualizer(modelo_agg, k = (2, 15), metric = 'calinski_harabasz')

visual_agg.fit(X)

visual_agg.poof();
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
kmeans = KMeans(n_clusters = 2)

kmeans.fit(X)

X['labels'] = kmeans.labels_
```

```python
pd.plotting.scatter_matrix(X, c = X['labels'], cmap = 'viridis');
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
pca = PCA().fit(X)

pca_x = pca.transform(X)

pca_x.shape
```

```python
fig = plt.figure(facecolor = 'none', figsize = (10, 10))


colors = np.where((X['labels'] >= 2) & (X['labels'] <= 3), 'red', 'blue')

ax = plt.subplot2grid((6, 1), (0, 0), rowspan = 4, facecolor = 'none')

ax.scatter(pca_x[:,0], pca_x[:,1], c = X['labels'])

ax.set_xlabel('PCA 1')
ax.set_ylabel('PCA 2')

y = np.ones(X.shape[0])
ax.tick_params(axis = 'both', colors = 'grey')
ax.xaxis.label.set_color('grey')
ax.yaxis.label.set_color('grey')
ax.spines['bottom'].set_color('grey')
ax.spines['top'].set_color('grey')
ax.spines['right'].set_color('grey')
ax.spines['left'].set_color('grey')

plt.show();
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
fig, ax = plt.subplots(1, 4, figsize=(15, 7), facecolor = 'none')

for i, p in enumerate([5, 30, 50, 100]):
    
    tsne = TSNE(perplexity = p)
    
    x_emb = tsne.fit_transform(X)
    
    ax[i].set_facecolor('none')
    ax[i].scatter(x_emb[:,0], x_emb[:,1], c = X['labels'])
    ax[i].set_title('Perplexity = {}'.format(p), color = 'grey')
    ax[i].tick_params(axis = 'both', colors = 'grey')
    ax[i].xaxis.label.set_color('grey')
    ax[i].yaxis.label.set_color('grey')
    ax[i].spines['bottom'].set_color('grey')
    ax[i].spines['top'].set_color('grey')
    ax[i].spines['right'].set_color('grey')
    ax[i].spines['left'].set_color('grey')
    
plt.show();
```
