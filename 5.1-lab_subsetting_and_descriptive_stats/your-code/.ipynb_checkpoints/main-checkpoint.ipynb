#### Find the minimum, mean and the maximum of all numeric columns for each Department.

##### Hint: Use functions from Data Manipulation lesson

FUNCTION TO DROP MULTI-INDEX COLUMNS PIVOT

```python
#df.groupby("Department").describe()[["min", "mean","max"]]

sumdf=df.pivot_table(index=["Department"], values=["Salary"], aggfunc="describe")
sumdf.columns = sumdf.columns.droplevel(level=0)
sumdf=sumdf[["min","mean", "max"]]
```

### Bonus Question

#### For each department, compute the difference between the maximal salary and the minimal salary.

##### * Hint: try using `agg` or `apply` and `lambda`*

```python
# Apply the custom function to each row and create a new 'difference' column
display(sumdf)

sumdf.apply(lambda row: row['max'] - row['min'], axis=1)
```
